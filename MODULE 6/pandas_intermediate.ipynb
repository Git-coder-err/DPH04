## **Cleaning Data**

Cleaning data involves eliminating or rectifying inaccuracies, inconsistencies, 
and missing values within your dataset, utilizing techniques such as handling 
missing values via deletion or imputation, rectifying data types, and detecting 
and eliminating duplicate entries, ultimately resulting in more precise and dependable analysis.

```python
# Objective Number 1: Turn the '?' into a null value
df.replace('?', pd.NA, inplace=True)
df
```

```python
# Objective Number 1: Turn the '?' into a null value
df.replace('?', pd.NA, inplace=True)
df
```

```python
# Replace null with unemployed
df["occupation"] = df["occupation"].fillna("Unemployed")
df
```

```python
# Checking every column for null values
df.isnull().sum()
```

```python
# Drop the rows that have null values
df.dropna(inplace=True)
df
```

### **Trimming and Cleaning Text Data**

```python
# Removing the spaces at the beginning and at the end of the string.
df['workclass'] = df["workclass"].str.strip()
df['workclass']
```

```python
occupation_mapping = {
    'Machine-op-inspct': 'Machine Operator',
    'Farming-fishing': 'Farming and Fishing',
    'Protective-serv': 'Protective Services'
}

df['occupation'] = df['occupation'].map(occupation_mapping).fillna(df['occupation'])
```

```python
df
```

```python
# Removing the hyphen and replace into space
df['marital-status'] = df['marital-status'].replace('-', ' ', regex=True)
```

```python
df
```

```python
# Change the data type of income column
df['income'] = df['income'].astype('category')
```

```python
df.dtypes
```

## **Handling Duplicates**

Identifying and removing duplicate records are crucial for maintaining data quality. 
Pandas provides .duplicated() and .drop_duplicates() for finding and removing duplicates, 
ensuring each data point is unique for accurate analysis.

```python
# True = Duplicated, False = Unique
df.duplicated()
```

```python
print("Duplicated Rows:", df.duplicated().sum())
```

```python
# Delete the duplicate rows
df.drop_duplicates(inplace=True)
df
```

```python
print("Duplicated Rows:", df.duplicated().sum())
```

### **Aggregating Data** (.groupby)

Aggregating data involves summarizing data points into meaningful statistics, 
such as averages, sums, or counts, which can be achieved using GroupBy operations or pivot tables. 
This helps in understanding the dataset at a higher level.

```python
df.columns
```

```python
df['occupation'].unique()
```

```python
# Getting the average age for each occupation.

age_series = df.groupby('occupation')['age'].mean()
age_series
```

```python
# Apply a function to the age column
def age_squared(age):
    return age ** 2
age_squared_series = df['age'].apply(age_squared)
age_squared_series
```
